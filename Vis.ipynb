```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data/EXINUS.csv', index_col=0)
fig,ax=plt.subplots(figsize=(8,5), dpi=100)
labels=['1994','1998','2002','2006','2010','2014','2018']
xt=[0,50,100,150,200,250, 300]
plt.plot(np.array(df['EXINUS']), linewidth=2)
ax.set_facecolor('#f1f0f0')
ax.grid()
ax.set_xticks(xt)
ax.set_xticklabels(labels)
ax.set_title('USD/INR Exchange Rate')
ax.set_xlabel('Year')
plt.show()
```
